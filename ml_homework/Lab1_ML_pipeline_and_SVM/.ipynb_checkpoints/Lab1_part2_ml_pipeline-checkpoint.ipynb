#### 2.3. PCA trasformation

Выберем первые восемь компонент, так как в таком случае суммарная доля объясняемой дисперсии будет больше 0.95

Метод `PCA` позволяет найти базис из собственных векторов в k-мерном пространстве, проекция выборки на который максимизирует суммарную дисперсию ($\:tr(\frac1n Y^TY)\:$). Если не нормализовать данные, то может существовать некоторый признак, у которого будет огромная дисперсия в виду его масштаба, и `PCA` будет стремиться проецировать данные по направлению этого признака и тем самым минимизировать вклад остальных признаков. Однако при нормировке может оказаться так, что найдётся другой признак с большей дисперсией, чем у исходного, и таким образом была бы потеряна существенная информация о выборке

```python
pca = PCA(n_components=8, random_state=seed)
```

#### 2.6. Bagging.
Here starts the ensembling part.

First we will use the __Bagging__ approach. Build an ensemble of $N$ algorithms varying N from $N_{min}=2$ to $N_{max}=100$ (with step 5).

We will build two ensembles: of logistic regressions and of decision trees.

*Comment: each ensemble should be constructed from models of the same family, so logistic regressions should not be mixed up with decision trees.*


*Hint 1: To build a __Bagging__ ensebmle varying the ensemble size efficiently you might generate $N_{max}$ subsets of `train` data (of the same size as the original dataset) using bootstrap procedure once. Then you train a new instance of logistic regression/decision tree with optimal hyperparameters you estimated before on each subset (so you train it from scratch). Finally, to get an ensemble of $N$ models you average the $N$ out of $N_{max}$ models predictions.*

*Hint 2: sklearn might help you with this taks. Some appropriate function/class might be out there.*

* Plot `f1` and `accuracy` scores plots w.r.t. the size of the ensemble.

* Briefly analyse the plot. What is the optimal number of algorithms? Explain your answer.

* How do you think, are the hyperparameters for the decision trees you found in 2.5 optimal for trees used in ensemble? 

```python
Nmin = 2
Nmax = 100
step = 5

# Generate bootstrap samples
n_train_obj, n_feat = X_train.shape
bs_indices = np.random.randint(low=0, high=n_train_obj, size=Nmax*n_train_obj)
bs_X_train = X_train_norm[bs_indices].reshape(Nmax, n_train_obj, n_feat)
bs_y_train = y_train_ie[bs_indices].reshape(Nmax, -1)

# Building ensembles
trees = []
log_regressions = []

for i in range(Nmax):
    tree = DecisionTreeClassifier(max_depth=8, random_state=seed)
    tree.fit(bs_X_train[i], bs_y_train[i])
    trees.append(tree)
    
    classifier = LogisticRegression(multi_class='multinomial', solver='saga', tol=1e-3, 
                                    C=10, penalty='l2', random_state=seed)
    classifier.fit(bs_X_train[i], bs_y_train[i])
    log_regressions.append(classifier)
```

```python
# Ensembles predictions
trees_pred = []
log_regr_pred = []

for i in range(Nmax):
    trees_pred.append(trees[i].predict(X_test_norm))    
    log_regr_pred.append(log_regressions[i].predict(X_test_norm))
    
trees_pred = np.array(trees_pred)
log_regr_pred = np.array(log_regr_pred)
```

```python
from scipy.stats import mode


# Get ensembles estimation
trees_acc = []
trees_f1 = []

lr_acc = []
lr_f1 = []

for N in range(Nmin, Nmax, step):
    N_trees_pred = mode(trees_pred[:N], axis=0)[0][0]
    trees_acc.append(accuracy_score(N_trees_pred, y_test_ie))
    trees_f1.append(f1_score(N_trees_pred, y_test_ie, average='weighted'))
    
    N_lr_pred = mode(log_regr_pred[:N], axis=0)[0][0]
    lr_acc.append(accuracy_score(N_lr_pred, y_test_ie))
    lr_f1.append(f1_score(N_lr_pred, y_test_ie, average='weighted'))
```

```python
xticks = np.arange(Nmin, Nmax, step)

plt.figure(figsize=(9, 5))
plt.plot(xticks, trees_acc, label='Accuracy')
plt.plot(xticks, trees_f1, label='F1 score')

plt.title('Trees ensemble', {'fontsize': 14})
plt.xticks(xticks)
plt.xlabel('N')
plt.grid(True)
plt.legend(fontsize='large')
plt.show()
```

Оптимальное количество деревьев с `max_depth=8` равно 97. 

Однако уже при 17 деревьях качество ансамбля близко к наилучшему

Деревья с ограниченной глубиной не совсем подходят для ансамбля, поскольку (в общем случае) этот фактор не позволяет им должным образом переобучаться, в результате чего `variance` перестаёт существенно уменьшаться

**Замечание**: В данном случае размер обучающей выборки порядка 500, а максимальное количество листов в дереве составляет 256, так что деревья будут почти идентичны деревьям с неограниченной глубиной

```python
xticks = np.arange(Nmin, Nmax, step)

plt.figure(figsize=(9, 5))
plt.plot(xticks, lr_acc, label='Accuracy')
plt.plot(xticks, lr_f1, label='F1 score')

plt.title('Logistic regressions ensemble', {'fontsize': 14})
plt.xticks(xticks)
plt.xlabel('N')
plt.grid(True)
plt.legend(fontsize='large')
plt.show()
```

Оптимальное количество линейных регрессоров равно 17. 

Далее с ростом числа базовых алгоритмов качество ансамбля варьируется и затем немного падает.
Связано это с переобучением модели (хоть и незначительным в силу использования бутстрапных выборок) и её простотой

#### 2.8. Learning curve
Your goal is to estimate, how does the model behaviour change with the increase of the `train` dataset size.

* Split the training data into 10 equal (almost) parts. Then train the models from above (Logistic regression, Desicion Tree, Random Forest) with optimal hyperparameters you have selected on 1 part, 2 parts (combined, so the train size in increased by 2 times), 3 parts and so on.

* Build a plot of `accuracy` and `f1` scores on `test` part, varying the `train` dataset size (so the axes will be score - dataset size.

* Analyse the final plot. Can you make any conclusions using it? 

```python
import math
```

```python
y_train = pd.get_dummies(y_train).values.argmax(axis=1)
y_test = pd.get_dummies(y_test).values.argmax(axis=1)

acc_res = []
f1_res = []
train_sizes = np.linspace(0.1, 1, 10)

for train_size_ratio in train_sizes:
    tmp_size = math.ceil(train_size_ratio*X_train.shape[0])
    X_train_tmp = X_train[0:tmp_size]
    y_train_tmp = y_train[0:tmp_size]
    
    log_regr = LogisticRegression(multi_class='multinomial', 
                                    solver='saga', 
                                    tol=1e-3, 
                                    C=10, 
                                    penalty='l2',  
                                    random_state=seed)
    tree = DecisionTreeClassifier(random_state=seed, max_depth=8)
    rand_forest = RandomForestClassifier(max_features='sqrt', 
                                         n_estimators=17,
                                         random_state=seed)
    
    scaler = StandardScaler()
    X_train_norm = scaler.fit_transform(X_train_tmp)
    X_test_norm = scaler.transform(X_test)
    
    log_regr.fit(X_train_norm, y_train_tmp)
    tree.fit(X_train_tmp, y_train_tmp)
    rand_forest.fit(X_train_tmp, y_train_tmp)
    
    acc_res.append(accuracy_score(log_regr.predict(X_test_norm), y_test))
    acc_res.append(accuracy_score(tree.predict(X_test), y_test))
    acc_res.append(accuracy_score(rand_forest.predict(X_test), y_test))
    
    f1_res.append(f1_score(log_regr.predict(X_test_norm), y_test, average='weighted'))
    f1_res.append(f1_score(tree.predict(X_test), y_test, average='weighted'))
    f1_res.append(f1_score(rand_forest.predict(X_test), y_test, average='weighted'))
```

```python
acc_res = np.array(acc_res).reshape(3, -1)
f1_res = np.array(f1_res).reshape(3, -1)

fig, axes = plt.subplots(1, 2, figsize=(13, 6))

axes[0].plot(train_sizes, acc_res[0], c='b', 
            label='Logistic regression')
axes[0].plot(train_sizes, acc_res[1], c='g', 
            label='Decision tree')
axes[0].plot(train_sizes, acc_res[2], c='orange', 
            label='Random forest')

axes[1].plot(train_sizes, f1_res[0], c='b', 
            label='Logistic regression')
axes[1].plot(train_sizes, f1_res[1], c='g', 
            label='Decision tree')
axes[1].plot(train_sizes, f1_res[2], c='orange', 
            label='Random forest')

axes[0].grid(True)
axes[0].legend()
axes[0].set_title('Accuracy score')
axes[0].set_xlabel('Train size')
axes[0].set_ylabel('Score')

axes[1].grid(True)
axes[1].legend()
axes[1].set_title('F1 score')
axes[1].set_xlabel('Train size')
axes[1].set_ylabel('Score')

plt.show()
```

**Вывод:** Поскольку гиперпараметры для вышеприведённых моделей были выбраны исходя из всей обучающей выборки, то логично, что они являются неоптимальными для обучающих выборок меньшего размера. Пожалуй, это является основной причиной резких скачков на графике.  
Если рассматривать алгоритмы по отдельности, то у лог.регрессии низкий score при малых значениях в силу неточности оптимизации и того факта, что она не смогла выявить всех зависимостей в данных.  
Решающее дерево переобучается при малых размерах данных, так же, как и случайный лес.
