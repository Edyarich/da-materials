__Задание 2 (1 балл):__ Опишите класс комплексных чисел. У пользователя должна быть возможность создать его объект на основе числа и в алгебраической форме, и в полярной. Класс должен поддерживать основные математические операции (+, -, \*, /) за счет перегрузки соответствующих магических методов. Также он должен поддерживать возможность получить число в алгебраической и полярной форме. Допускается использование модуля `math`.

````python
class ComplexNumber:
    EPS = 1e-6
    NUMERIC_TYPES = (float, int, bool)
    
    @staticmethod
    def _is_equal(first_num, second_num):
        input_types = set((type(first_num), type(second_num)))
        
        if not input_types.issubset(ComplexNumber.NUMERIC_TYPES):
            raise TypeError
            
        return abs(first_num - second_num) < ComplexNumber.EPS
    
    def __init__(self, first_num, second_num, form='algebra'):
        assert form in {'algebra', 'polar'}
        
        if form == 'algebra':
            self._init_alg_num(first_num, second_num)
            
        elif form == 'polar':
            self._init_polar_num(first_num, second_num)
    
    def _init_alg_num(self, x, y):
        self._x = x
        self._y = y
        
    def _init_polar_num(self, rho, phi):
        assert rho >= 0
        
        self._x = rho * math.cos(phi)
        self._y = rho * math.sin(phi)
    
    def get_x(self):
        return self._x
    
    def get_y(self):
        return self._y
    
    def get_alg_num(self):
        return self.get_x(), self.get_y()
    
    def get_rho(self):
        return (self.get_x()**2 + self.get_y()**2)**0.5
    
    def get_phi(self):
        return math.atan2(self.get_y(), self.get_x())
    
    def get_polar_form(self):
        return self.get_rho(), self.get_phi()
    
    def __eq__(self, oth_complex_num):
        obj_type = type(oth_complex_num)

        if obj_type in ComplexNumber.NUMERIC_TYPES:
            return ComplexNumber._is_equal(self.get_x(), oth_complex_num) and \
                   ComplexNumber._is_equal(self.get_y(), 0)
        
        elif obj_type == ComplexNumber:
            return ComplexNumber._is_equal(self.get_x(), oth_complex_num.get_x()) and \
                   ComplexNumber._is_equal(self.get_y(), oth_complex_num.get_y())
        
        else:
            return False
        
    def __add__(self, oth_complex_num):
        obj_type = type(oth_complex_num)
        
        if obj_type in ComplexNumber.NUMERIC_TYPES:
            sum_x = self.get_x() + oth_complex_num
            sum_y = self.get_y()
            
            return ComplexNumber(sum_x, sum_y)
        
        elif obj_type == ComplexNumber:
            sum_x = self.get_x() + oth_complex_num.get_x()
            sum_y = self.get_y() + oth_complex_num.get_y()
            
            return ComplexNumber(sum_x, sum_y)
        
        else:
            raise TypeError(f"unsupported operand type for +: {obj_type}")
    
    def __radd__(self, oth_complex_num):
        return self.__add__(oth_complex_num)
    
    def __neg__(self):
        return ComplexNumber(-self.get_x(), -self.get_y())
    
    def __sub__(self, other_complex_num):
        if not hasattr(other_complex_num, '__neg__'):
            raise TypeError(f"bad operand type for unary -:" + 
                            f"{type(other_complex_num)}")
        
        return self.__add__(-other_complex_num)
    
    def __rsub__(self, oth_complex_num):
        return -self.__sub__(oth_complex_num)
    
    def __mul__(self, oth_complex_num):
        obj_type = type(oth_complex_num)
        
        if obj_type in ComplexNumber.NUMERIC_TYPES:
            product_x = self.get_x() * oth_complex_num
            product_y = self.get_y() * oth_complex_num
            
            return ComplexNumber(product_x, product_y)
        
        elif obj_type == ComplexNumber:
            product_rho = self.get_rho() * oth_complex_num.get_rho()
            product_phi = self.get_phi() + oth_complex_num.get_phi()
            
            return ComplexNumber(product_rho, product_phi, 'polar')
        
        else:
            raise TypeError(f"unsupported operand type for *: {obj_type}")
            
    def __rmul__(self, oth_complex_num):
        return self.__mul__(oth_complex_num)
    
    def __truediv__(self, oth_complex_num):
        if oth_complex_num == 0:
            raise ZeroDivisionError
        
        obj_type = type(oth_complex_num)
        
        if obj_type in ComplexNumber.NUMERIC_TYPES:
            quotient_x = self.get_x() / oth_complex_num
            quotient_y = self.get_y() / oth_complex_num
            
            return ComplexNumber(quotient_x, quotient_y)
        
        elif obj_type == ComplexNumber:
            quotient_rho = self.get_rho() / oth_complex_num.get_rho()
            quotient_phi = self.get_phi() - oth_complex_num.get_phi()
            
            return ComplexNumber(quotient_rho, quotient_phi, 'polar')
        
        else:
            raise TypeError(f"unsupported operand type for /: {obj_type}")
            
    def __rtruediv__(self, oth_complex_num):
        return ComplexNumber(1, 0).__truediv__(self.__truediv__(oth_complex_num))
    
    def __repr__(self):
        if self.get_y() >= 0:
            return f"{self.get_x()} + {self.get_y()}i"
        
        else:
            return f"{self.get_x()} - {abs(self.get_y())}i"
````

__Задание 3 (2 балла):__ Опишите класс для векторов в N-мерном пространстве. В качестве основы  используйте список значений координат вектора, задаваемый `list`. Обеспечьте поддержку следующих операций: сложение, вычитание (с созданием нового вектора-результата), скалярное произведение, косинус угла, евклидова норма. Все операции, которые можно перегрузить с помощью магических методов, должны быть реализованы именно через них. Класс должен производить проверку консистентности аргументов для каждой операции и в случаях ошибок выбрасывать исключение `ValueError` с исчерпывающим объяснением ошибки.

````python
class Vector:
    EPS = 1e-6
    NUMERIC_TYPES = (float, int, bool)
    
    @staticmethod
    def _is_equal(first_num, second_num):
        input_types = set((type(first_num), type(second_num)))
        
        if not input_types.issubset(Vector.NUMERIC_TYPES):
            raise TypeError
            
        return abs(first_num - second_num) < Vector.EPS
    
    def __init__(self, vector_values_list):
        is_num_type = list(map(lambda elem: type(elem) in Vector.NUMERIC_TYPES, 
                               vector_values_list))
        
        if not all(is_num_type):
            raise ValueError("the vector contains non-numeric data")
            
        self._dim = len(vector_values_list)
        self._vec = vector_values_list
        
    @property
    def vec(self):
        return self._vec
    
    @property
    def dim(self):
        return self._dim
    
    def __eq__(self, other):
        other_type = type(other)

        if other_type in Vector.NUMERIC_TYPES:
            is_equal_vec = list(map(lambda elem: Vector._is_equal(elem, other), 
                                    self.vec))
            
            return all(is_equal_vec)
        
        elif other_type == Vector:
            if other.dim != self.dim:
                return False
                
            is_equal_vec = [Vector._is_equal(first_elem, second_elem) 
                            for first_elem, second_elem in zip(self.vec, other.vec)]
            
            return all(is_equal_vec)
        
        else:
            raise TypeError(f"unsupported operand type for ==: {other_type}")       
    
    def __add__(self, other):
        other_type = type(other)

        if other_type in Vector.NUMERIC_TYPES:
            sum_vector = list(map(lambda elem: elem + other, 
                                  self.vec))
            
            return Vector(sum_vector)
        
        elif other_type == Vector:
            if other.dim != self.dim:
                raise ValueError("different vectors sizes for +")
                
            sum_vector = [first_elem + second_elem 
                          for first_elem, second_elem in zip(self.vec, other.vec)]
            
            return Vector(sum_vector)
        
        else:
            raise TypeError(f"unsupported operand type for +: {other_type}") 
            
    def __radd__(self, other):
        return self.__add__(other)
    
    def __neg__(self):
        return Vector([-elem for elem in self.vec])
    
    def __sub__(self, other):
        if not hasattr(other, '__neg__'):
            raise TypeError(f"bad operand type for unary -: {type(other)}")
            
        return self.__add__(-other)
    
    def __rsub__(self, other):
        return -self.__sub__(other)
    
    def dot(self, other):
        if type(other) != Vector:
            raise TypeError(f"bad operand type for 'dot': {type(other)}")
            
        if other.dim != self.dim:
            raise ValueError("different vectors sizes for 'dot'")
            
        return sum([first_elem * second_elem for 
                    first_elem, second_elem in zip(self.vec, other.vec)])
    
    def l2_norm(self):
        return sum([elem**2 for elem in self.vec])**0.5
    
    def angle(self, other):
        if type(other) != Vector:
            raise TypeError(f"bad operand type for 'dot': {type(other)}")
            
        if other.dim != self.dim:
            raise ValueError("different vectors sizes for 'dot'")
            
        self_norm = self.l2_norm()
        other_norm = other.l2_norm()
        scalar_product = self.dot(other)
        
        if Vector._is_equal(self_norm, 0) or Vector._is_equal(other_norm, 0):
            return 0
        
        return math.acos(scalar_product / self_norm / other_norm)
    
    def __repr__(self):
        return f"{self.vec}"
````

__Задание 4 (2 балл):__ Опишите декоратор, который принимает на вход функцию и при каждом её вызове печатает строку "This function was called N times", где N - число раз, которое это функция была вызвана на текущий момент (пока функция существует как объект, это число, очевидно, может только неубывать).

````python
def calls_counter(func):
    calls_count = 0
    
    def __calls_counter(*args, **kwargs):
        result = func(*args, **kwargs)
        
        nonlocal calls_count
        calls_count += 1
        
        print(f"This function was called {calls_count} times")
        
        return result
    
    return __calls_counter
````

Tests for `calls_counter` decorator

````python
@calls_counter
def fib(i: int) -> int:
    if i < 0:
        raise ValueError("The index i must be greater or equal than zero")
    
    elif i == 0:
        return 0
    
    elif i == 1:
        return 1
    
    return fib(i-1) + fib(i-2)
````

````python
with open("tmp_out.txt", "w") as out_fd:
    with redirect_stdout(out_fd):
        fib(0) # Calls fib 1 time
        fib(1) # Calls fib 1 time
        fib(2) # Calls fib 3 times: fib(2), fib(1), fib(0)
        
with open("tmp_out.txt", "r") as in_fd:
    calls_count = 0
    
    for i, line in enumerate(in_fd.readlines()):
        calls_count += 1
        print(line.strip() == f"This function was called {i+1} times")
        
    print(calls_count == 5)
    
with open("tmp_out.txt", "w") as out_fd:
    with redirect_stdout(out_fd):
        fib(3) # Calls fib 5 times: fib{3}, fib(1), fib(2), fib(1), fib(0)
        
with open("tmp_out.txt", "r") as in_fd:
    calls_count = 0
    
    for i, line in enumerate(in_fd.readlines()):
        calls_count += 1
        print(line.strip() == f"This function was called {i+6} times")
        
    print(calls_count == 5)
    
os.remove("tmp_out.txt")
````

__Задание 5 (3 балла):__ Опишите декоратор класса, который принимает на вход другой класс и снабжает декорируемый класс всеми атрибутами входного класса, названия которых НЕ начинаются с "\_". В случае конфликтов имён импортируемый атрибут должен получить имя с суффиксом "\_new".

````python
def copy_class_attrs(cls_giver):
    def __copy_class_attrs(updating_cls):
        for new_attr in dir(cls_giver):
            if new_attr.startswith('_'):
                continue
            
            new_attr_name = new_attr
            if hasattr(updating_cls, new_attr):
                new_attr_name += '_new'
            
            setattr(updating_cls, new_attr_name, getattr(cls_giver, new_attr))
            
        return updating_cls
    
    return __copy_class_attrs
````

Tests for `copy_class_attrs` decorator

````python
class A:
    name = 'Ivan'
    surname = 'Ivanov'
    email = "iv***@gmail.com"
    
    def __init__(self, name, surname, email):
        self.name = name
        self.surname = surname
        self.email = email
        
    def foo(self):
        print(email)
    
    def bar(self):
        return name + ' ' + surname

@copy_class_attrs(A)
class B:
    name = 'Semen'
    surname = 'Semenov'
    mobile_phone = "8800*****35"
    
    def __init__(self, name, surname, phone):
        self.name = name
        self.surname = surname
        self.mobile_phone = phone
        
    def foo(self):
        print(mobile_phone)
    
    def other_bar(self):
        pass
    
B_attrs = list(filter(lambda attr: not attr.startswith('_'), dir(B)))
old_attrs = ['name', 'surname', 'mobile_phone', 'foo', 'other_bar']
new_attrs = ['foo_new', 'bar', 'name_new', 'surname_new', 'email']

print([attr in B_attrs for attr in old_attrs])
print([attr in B_attrs for attr in new_attrs])
````

__Задание 7 (5 баллов):__ Ставится задача расчета стоимости чашки кофе. Опишите классы нескольких типов кофе (латте, капучино, американо), а также классы добавок к кофе (сахар, сливки, кардамон, пенка, сироп). Используйте шаблон "декоратор". Каждый класс должен характеризоваться методом вычисления стоимости чашки `calculate_cost`. Пользователь должен иметь возможность комбинировать любое число добавок с выбранным кофе и получить на выходе общую стоимость:

```
Cream(Sugar(Latte())).calculate_cost()
```

Первым элементом чашки всегда должен быть сам кофе, а не добавка, в противном случае при попытке создания чашки должно выбрасываться исключение:

```
Cream(Latte(Sugar())).calculate_cost() -> exception
```

Кофе может встречаться в чашке только один раз, в противном случае при попытке создания чашки должно выбрасываться исключение:

```
Cappuccino(Sugar(Latte())).calculate_cost() -> exception
```

Добавки могут включаться в чашку в любом количестве и порядке.
Добавление новых типов кофе и добавок не должно требовать изменения существующего кода.

````python
LATTE_COST = 160
CAPPUCCINO_COST = 170
AMERICANO_COST = 150

SUGAR_COST = 35
CREAM_COST = 20
CARDAMOM_COST = 25
FOAM_COST = 5
SYRUP_COST = 15
````

````python
# Interface of coffee drink
class CoffeeInterface:
    def calculate_cost(self): raise NotImplementedError

# Main drinks
class Latte(CoffeeInterface):
    def calculate_cost(self): 
        return LATTE_COST
    
class Cappuccino(CoffeeInterface):
    def calculate_cost(self): 
        return CAPPUCCINO_COST
    
class Americano(CoffeeInterface):
    def calculate_cost(self): 
        return AMERICANO_COST

# Topics
class Sugar(CoffeeInterface):
    def __init__(self, coffee):
        self.coffee = coffee
        
    def calculate_cost(self):
        return SUGAR_COST + self.coffee.calculate_cost()
    
class Cream(CoffeeInterface):
    def __init__(self, coffee):
        self.coffee = coffee
        
    def calculate_cost(self):
        return CREAM_COST + self.coffee.calculate_cost()
    
class Cardamom(CoffeeInterface):
    def __init__(self, coffee):
        self.coffee = coffee
        
    def calculate_cost(self):
        return CARDAMOM_COST + self.coffee.calculate_cost()
    
class Foam(CoffeeInterface):
    def __init__(self, coffee):
        self.coffee = coffee
        
    def calculate_cost(self):
        return FOAM_COST + self.coffee.calculate_cost()
    
class Syrup(CoffeeInterface):
    def __init__(self, coffee):
        self.coffee = coffee
        
    def calculate_cost(self):
        return SYRUP_COST + self.coffee.calculate_cost()
````

Tests for `CoffeeInterface` implementations

````python
print(Latte().calculate_cost() == LATTE_COST)
print(Cappuccino().calculate_cost() == CAPPUCCINO_COST)
print(Cardamom(Americano()).calculate_cost() == AMERICANO_COST+CARDAMOM_COST)
print(Syrup(Foam(Cappuccino())).calculate_cost() == SYRUP_COST+FOAM_COST+CAPPUCCINO_COST)
print(Cardamom(Cream(Sugar(Latte()))).calculate_cost() == CARDAMOM_COST+CREAM_COST+SUGAR_COST+LATTE_COST)
print(Foam(Foam(Sugar(Foam(Americano())))).calculate_cost() == AMERICANO_COST+SUGAR_COST+FOAM_COST*3)

try:
    print(Sugar(Syrup()).calculate_cost())
except TypeError:
    print("OK")
    
try:
    print(Latte(Sugar(Cappuccino())).calculate_cost())
except TypeError:
    print("OK")
````
