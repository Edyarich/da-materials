```python
import numpy as np
import scipy.linalg as linalg
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
```

```python
A = np.array([[1, 1], [1, 15]], dtype=int)
b = np.array([f(1), f(15)])
res = linalg.solve(A, b)
```

```python
C = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]], dtype=int)
d = np.array([f(1), f(8), f(15)])
sq_res = linalg.solve(C, d)
```

```python
E = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]], dtype=int)
f = np.array([f(1), f(4), f(10), f(15)])
cb_res = linalg.solve(E, f)
```

```python
dots = np.linspace(0, 15, 150)

plt.figure(figsize=(10, 6))
plt.plot(dots, f(dots), color='r', label='f(x)')
plt.plot(dots, res[0] + dots * res[1], color='b', label='line')
plt.plot(dots, sq_res[0] + dots * sq_res[1] + dots**2 * sq_res[2], 
         color='g', label='square_line')
plt.plot(dots, cb_res[0] + dots * cb_res[1] + dots**2 * cb_res[2] + dots**3 * cb_res[3], 
         color='yellow', label='coub_line')
plt.xticks(np.arange(0, 16))
plt.title('  f(x)   ', fontsize=20)
plt.grid(ls=':')
plt.legend(fontsize=16)
plt.show()
```
