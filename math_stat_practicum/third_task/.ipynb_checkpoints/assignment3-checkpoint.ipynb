Сгенерируйте выборку из трех распределений: нормального, Лапласа и Коши размера `N=500`.

Подумайте для каких распределений сложнее найти значение математического ожидания по выборке? Почему?

**Ответ:** У распределения Коши не существует мат.ожидания, а у распределения Лапласа довольно "тяжёлый" хвост, так что потребуется большое количество измерений для нивелирования выбросов. Поэтому, легче всего определить мат.ожидание у нормального распределения

Часто, в реальной жизни, мы сталкиваемся с данными в которых есть выбросы возникшие, например, из-за опечаток. Пусть в выборке содержится 99 точек на интервале $[0;1]$ и один элемент равный 300. Тогда среднее значение нашей выборки будет не меньше трех. Ясно, что так дело не пойдет, что же делать?

Можно использовать медиану или **усеченное среднее**.

**Усеченное среднее порядка $\alpha$** это среднее составленное из выборки от которой отрезаны $\alpha$ и $1 - \alpha$ квантили. Как ее получить?


1.   Отсортировать выборку
2.   Отрезать первые и последние ($\alpha \cdot \text{размер выборки}$) элементов выборки
3.   Посчитать среднее полученной выборки

Напишите функцию для подсчета этой статистики.

```python
import math

def truncated_mean(sample, alpha=0.3):    
    first_ind = math.floor(alpha * sample.size)
    last_ind = math.ceil((1 - alpha) * sample.size)
               
    sorted_sample = np.sort(sample)
    cut_sorted_sample = sorted_sample[first_ind:last_ind]
                         
    return cut_sorted_sample.mean()
```

**Теорема**
Пусть распределение симметрично относительно нуля и имеет одну моду в нуле, тогда асимптотическая дисперсия усеченного среднего порядка $\alpha$ может быть выражена по формуле:
$$
\sigma^2(\alpha) = 2\cdot\frac{\int_0^{u_{1 - \alpha}} x^2 dF(x) + \alpha\cdot(u_{1 - \alpha})^2}{(1 - 2\alpha)^2},
$$

где $u_\alpha$ – квантиль порядка $\alpha$.

Напишите функцию, которая это делает.

*Подсказка.* Для численного интегрирования используйте функцию `quad` из модуля `scipy.integrate`

```python
from scipy.integrate import quad

def asymp_var(distr, alpha):
    if alpha == 0.5:
        return 0
    
    quantile = distr.ppf(1 - alpha)
    integral, _ = quad(lambda x: x * x * distr.pdf(x), 0, quantile)
    return 2 * (integral + alpha * np.square(quantile)) / np.square(1 - 2 * alpha)
```

Постройте графики для асимптотической дисперсии в зависимости от $\alpha$ для всех трех распределений на разных осях (используйте `plt.subplots(1, 3, figsize=(21,7))`). **Не забудьте про цвета**. Для заголовка графика используйте `ax[i].set_title`, для заголовка всей картинки используйте `fig.suptitle`.

```python
alpha_grid = np.linspace(0.05, 0.49999, 100)
norm_asymp_vars = list(map(lambda alpha: asymp_var(norm_distr, alpha), alpha_grid))
laplacian_asymp_vars = list(map(lambda alpha: asymp_var(laplacian_distr, alpha), alpha_grid))
cauchy_asymp_vars = list(map(lambda alpha: asymp_var(cauchy_distr, alpha), alpha_grid))

fig, axes = plt.subplots(1, 3, figsize=(21, 7))
axes[0].plot(alpha_grid, norm_asymp_vars, c=color_for_normal)
axes[1].plot(alpha_grid, laplacian_asymp_vars, c=color_for_laplace)
axes[2].plot(alpha_grid, cauchy_asymp_vars, c=color_for_cauchy)

axes[0].set_title("Normal asymp.vars")
axes[1].set_title("Laplacian asymp.vars")
axes[2].set_title("Cauchy asymp.vars")

axes[0].set_xlabel("Alphas")
axes[1].set_xlabel("Alphas")
axes[2].set_xlabel("Alphas")

axes[0].set_ylabel("Asymp.var")
axes[1].set_ylabel("Asymp.var")
axes[2].set_ylabel("Asymp.var")

fig.suptitle("Asymptotic variances of various distributions")
plt.show()
```

Как можно интепретировать такой график? Напишите развернутый вывод.

*Подсказка.* Подумайте, какой известной оценке соответствует усеченное среднее порядка 0.5? А порядка 0?

**Ответ:** Усечённое среднее порядка 0 соответствует среднему арифметическому, порядка 0.5 - медиане.  
$\sigma^2(0) = EX^2 ,\: \sigma^2(0.5) = \dfrac{1}{4p^2(0)}$ - асимптотическая дисперсия для выборочной медианы.  
Для распределений с большим хвостом и с отсутствующим вторым моментом асимптотическая дисперсия усечённого среднего будет уменьшаться с увеличением альфа. Однако для менее тяжеловесных распределений (таких как нормальное) рассматривать усечённое среднее будет не так эффективно, поскольку с увеличением альфа растёт и асимптотическая дисперсия

Для каждого распределение на отдельном графике постройте значение трех оценок: среднего, медианы и усученного среднего порядка $0.35$. Для усеченного среднего проведите линию пунктиром (`fmt="--"`), для медианы точечками (`fmt=":"`). **Не забудьте про цвета**. То есть цвета на одном графике одинаковые, разные оценки отличает только формат линии.

**Ниже посчитайте все оценки необходимые для всех распределений**

```python
N_arr = np.arange(1, N+1)

def get_means(sample):
    return np.cumsum(sample) / N_arr


def get_trimmed_means(sample, alpha=0.35):
    return list(map(lambda n: truncated_mean(sample[:n], alpha), N_arr))


def get_medians(sample):
    return list(map(lambda n: np.median(sample[:n]), N_arr))
```

**Нормальное распределение**

```python
norm_means = get_means(norm_sample)
norm_medians = get_trimmed_means(norm_sample)
norm_trimmed_means = get_medians(norm_sample)

plt.figure(figsize=(12, 7))
plt.plot(N_arr, norm_means, c=color_for_normal, lw=2, label="Mean")
plt.plot(N_arr, norm_medians, ':', c=color_for_normal, lw=2, label="Median")
plt.plot(N_arr, norm_trimmed_means, '--', c=color_for_normal, lw=2, label="Trimmed mean")
plt.ylim(-0.5, 0.1)
plt.xlabel("Sample size")
plt.legend()
plt.show()
```

**Распределение Лапласа**

```python
laplac_means = get_means(laplacian_sample)
laplac_medians = get_trimmed_means(laplacian_sample)
laplac_trimmed_means = get_medians(laplacian_sample)

plt.figure(figsize=(12, 7))
plt.plot(N_arr, laplac_means, c=color_for_laplace, lw=2, label="Mean")
plt.plot(N_arr, laplac_medians, ':', c=color_for_laplace, lw=2, label="Median")
plt.plot(N_arr, laplac_trimmed_means, '--', c=color_for_laplace, lw=2, label="Trimmed mean")
plt.ylim(-0.25, 0.1)
plt.xlabel("Sample size")
plt.legend()
plt.show()
```

**Распределение Коши**

```python
cauchy_means = get_means(cauchy_sample)
cauchy_medians = get_trimmed_means(cauchy_sample)
cauchy_trimmed_means = get_medians(cauchy_sample)

plt.figure(figsize=(12, 7))
plt.plot(N_arr, cauchy_means, c=color_for_cauchy, lw=2, label="Mean")
plt.plot(N_arr, cauchy_medians, ':', c=color_for_cauchy, lw=2, label="Median")
plt.plot(N_arr, cauchy_trimmed_means, '--', c=color_for_cauchy, lw=2, label="Trimmed mean")
plt.ylim(-2, 2)
plt.xlabel("Sample size")
plt.legend()
plt.show()
```

Как эти три графика соотносятся с графиками асимптотической дисперсии усеченного среднего для всех этих распределений? Сделайте вывод.

<i>Обозначение<i>: АД - асимптотическая дисперсия

**Вывод:**   
<i>Лаплас<i>: Поскольку АД усечённого среднего меньше дисперсии, то усечённое среднее будет в среднем менее отклонено от нуля, чем среднее арифметическое, так как у данного распределения "тяжёлые" хвосты. В свою очередь медиана в среднем будет ещё менее отклонено от нуля, чем усечённое среднее ввиду минимальности АД 

<i>Коши<i>:  В силу отсутствия первого момента на графике можно увидеть, что среднее арифметическое может быть сколь угодно большим по модулю - это соответсвует тому, что $\sigma^2(0) = +\infty$. А медиана и выборочное среднее прекрасно сходятся к нулю, причём для данного альфа АД усечённого среднего будет чуть меньше, чем АД медианы

Найдите ОМП для двух этих распределений в предположении, что обе выбоки из многомерного нормального распределения. Не мудрите, $\mu$ и $\Sigma$ получаются в одну команду через `numpy`. Далее получите плотности для сетки `X`, `Y`. Используйте `sps.multivariate_normal`.

Замечание: Исходя из правдивости данного источника https://xavierbourretsicotte.github.io/MLE_Multivariate_Gaussian.html, при вычислении матрицы ковариации нужно делить на количество объектов $m$, а не на $m-1$. Поэтому в вызове `np.cov` указан параметр `bias=True`, а `rowvar=False` указан в силу того, что переменные расположены по столбцам, а измерения - по строкам

```python
bad_vines = data.loc[data.quality == 3][['alcohol', 'citric acid']].values
good_vines = data.loc[data.quality == 8][['alcohol', 'citric acid']].values

bad_vines_mll_mean = bad_vines.mean(axis=0)
good_vines_mll_mean = good_vines.mean(axis=0)

bad_vines_mll_cov = np.cov(bad_vines, rowvar=False, bias=True)
good_vines_mll_cov = np.cov(good_vines, rowvar=False, bias=True)

bad_vines_pdf = sps.multivariate_normal(mean=bad_vines_mll_mean, 
                                        cov=bad_vines_mll_cov).pdf
good_vines_pdf = sps.multivariate_normal(mean=good_vines_mll_mean, 
                                        cov=good_vines_mll_cov).pdf
```

Нарисуйте график плотности с линиями контура и нанесите точки выборки. Используйте зеленый цвет для вина плохого качества (`get_density_cmap("Greens")` и `"g"` для точек) и синий для вина хорошего качества (`get_density_cmap("Blues")` и `"b"`). Используйте функции `plt.pcolormesh`, `plt.clabel` и `plt.scatter`. Не забудьте про `plt.xlabel`, `plt.ylabel` и `plt.legend`.

```python
from matplotlib.colors import ListedColormap

def get_density_cmap(name):
    cmap =  plt.get_cmap(name)
    cmap_col = cmap(np.arange(cmap.N))
    cmap_col[:, -1] = np.linspace(0, 1, cmap.N)**2
    return ListedColormap(cmap_col)
```

```python
alcohol_ticks = np.linspace(8, 14.1, 200)
citr_ac_ticks = np.linspace(-0.1, 0.75, 200)

X, Y = np.meshgrid(alcohol_ticks, citr_ac_ticks)
grid = np.vstack([np.ravel(X), np.ravel(Y)]).T

bw_levels = bad_vines_pdf(grid).reshape(200, 200)
gw_levels = good_vines_pdf(grid).reshape(200, 200)

plt.figure(figsize=(15, 7))

plt.scatter(bad_vines[:, 0], bad_vines[:, 1], c="g", label="Bad vines")
plt.scatter(good_vines[:, 0], good_vines[:, 1], c="b", label="Good vines")

fig1 = plt.pcolormesh(X, Y, bw_levels, cmap=get_density_cmap("Greens"), 
                        shading='auto')
fig2 = plt.pcolormesh(X, Y, gw_levels, cmap=get_density_cmap("Blues"), 
                        shading='auto')

plt.colorbar(fig1)
plt.colorbar(fig2)

plt.xlabel("Alcohol")
plt.ylabel("Citric acid")
plt.legend()
plt.show()
```

Что можно сказать о вине, которому сомелье дали наивысший балл по сравнению с вином, которому дали наименьший балл, основываясь на график выше?

**Ответ:** Основываясь на графике выше, можно заключить, что у хороших вин содержание алкоголя выше, чем у плохих вин, и что содержание лимонной кислоты в плохих винах в основном меньше, чем у хороших вин

# Задача 3

Рассмотрим $X_1, ..., X_n \sim Bern(\theta)$. По сетке значений $\theta \in
    [0, 1]$ с шагом 0.01 постройте график зависимости нижней оценки дисперсии
    произвольной несмещенной оценки из неравенства Рао-Крамера от $\theta$. 

**Нер-во Рао-Крамера**: $D_\theta \theta^* \geqslant \dfrac{1}{n \cdot i(\theta)} , \text{ где }$
                        $i(\theta) = \dfrac{1}{\theta (1 - \theta)}$
                        
$\overline{X} $ - эффективная оценка парамметра $\theta$

```python
thetas_grid = np.arange(0.01, 1, 0.01)
fisher_info = 1.0 / thetas_grid / (1 - thetas_grid)
n = 1000

plt.figure(figsize=(10, 6))
plt.plot(thetas_grid, fisher_info/n)
plt.xlabel("Thetas")
plt.ylabel("Values")
#plt.xlim(0.2, 0.8)
#plt.ylim(0, 0.02)
plt.title("Правая часть из неравенства Рао-Крамера")
plt.show()
```

Какой можно сделать вывод (напишите в комментариях)?

<i>Обозначение<i>: ДОО - дисперсия оптимальной оценки

**Вывод**: Из неравенства Рао-Крамера можно заключить, что ДОО линейно уменьшается с ростом размера выборки. А из графика выше можно понять, что если истинное значение параметра $\theta$ лежит в промежутке $(0.2, 0.8)$, то ДОО будет мала по сравнению с остальными значениями параметра, причем ДОО почти неотличаются друг от друга при $\theta \in (0.2, 0.8)$

 Для каждого значения $\theta$ (для той же сетки) сгенерируйте выборку размера $n=1000$ для параметра $\theta$, посчитайте эффективную оценку $\theta$ и бутстрепную оценку дисперсии (количество бутстрепных выборок равно $1000$) этой эффективной оценки $\theta$.

```python
bs_samples_cnt = 1000
bs_estimations = []
bs_estimations_std = []

for theta in thetas_grid:
    sample = sps.bernoulli(p=theta).rvs(size=n)
    # rao_estimations.append(sample.mean())
    
    bs_indices = np.random.randint(0, n, bs_samples_cnt*n)
    bs_samples = sample[bs_indices].reshape(bs_samples_cnt, n)
    bs_means = bs_samples.mean(axis=1)
    bs_estimations.append(bs_means.mean())
    bs_estimations_std.append(bs_means.std())
    
bs_estimations = np.array(bs_estimations)
bs_estimations_std = np.array(bs_estimations_std)
```

 Нарисуйте график зависимости полученных бутстрепных оценок от $\theta$.

```python
plt.figure(figsize=(12, 7))

plt.fill_between(thetas_grid, bs_estimations-bs_estimations_std, 
                    bs_estimations+bs_estimations_std, alpha=0.3, 
                    label="+-std_range")
plt.plot(thetas_grid, bs_estimations, c="b", label="Bootstrap estimation")
plt.plot(thetas_grid, thetas_grid, c="orange", label="Ideal estimation")
plt.xlabel("Thetas")
plt.ylabel("Estimation")
plt.legend()
plt.show()
```

**Вывод**: Эффективные оценки, построенные по бутстрепным выборкам, отлично приближают истинное значение $\theta$ с довольно маленьким стандартным отклонением
