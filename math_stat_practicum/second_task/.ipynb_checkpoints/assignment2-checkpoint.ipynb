```python
cauchy_sample = pd.read_csv("Cauchy.csv")["sample"].values
```

```python
GRID_STEP = 0.01
PARAM_RANGE = (-1000, 1000)
PARAM_GRID = np.arange(*PARAM_RANGE, GRID_STEP)

def likelihood_func(sample, x):
    return -np.sum(np.log(1 + np.square(sample - x)) + np.log(np.pi))


def get_max_ll_estim(sample):
    ll_func_values = np.array(list(map(
                                lambda val: likelihood_func(sample, val), 
                                PARAM_GRID)))
    return PARAM_GRID[ll_func_values.argmax()]
```

Проведем небольшой предварительный анализ. Итак, если наши данные распределены согласно распределению Вейбулла, то справедливы следующие рассуждения:
$$
\begin{aligned} F(x) &=1-e^{-(x)^{\gamma}} \\-\ln (1-F(x)) &=x^{\gamma} \\ \underbrace{\ln (-\ln (1-F(x)))}_{y^{\prime}} &=\underbrace{\gamma \ln x}_{\text {kx }^{\prime}} \end{aligned}
$$
А значит и
$$
\underbrace{\ln (-\ln (1-\hat{F}(x)))}_{y^{\prime}} ≈\underbrace{\gamma \ln x}_{\text {kx }^{\prime}}
$$


Однако здесь мы сталкиваемся с проблемой: в нашей выборке есть нули. А логарифм нуля, как известно, не определен. Мы предлагаем вам сравнить два способа рещения этой проблемы:

1. Отбросить нули до построения эмпирической функции распределения $\hat{F}(x)$
2. Построить эмпирическую функцию, а затем учитывать только те точки, x-координата которых положительна.

Для первого способа отфильтруйте выборку в ячейке ниже

```python
non_zero_sample = weibull_sample[weibull_sample > 0]
```

Для каждого способа подсчитайте эмпирическую функцию распределения и 
$$y\prime = \ln (-\ln (1-\hat{F}(x))) \quad x' = ln x$$
 где $x$ – элементы исходной выборки. 

```python
from statsmodels.distributions.empirical_distribution import ECDF

first_ecdf = ECDF(non_zero_sample)
snd_ecdf = ECDF(weibull_sample)

nonzero_ind = snd_ecdf.x.nonzero()
snd_ecdf.x = snd_ecdf.x[nonzero_ind]
snd_ecdf.y = snd_ecdf.y[nonzero_ind]
snd_ecdf.n = nonzero_ind[0].size
```

```python
def modified_ecdf(ecdf, x):
    return np.log(-np.log(1 - ecdf(x)))


def modified_arg(x):
    return np.log(x)


mod_first_ecdf_val = modified_ecdf(first_ecdf, non_zero_sample)
mod_second_ecdf_val = modified_ecdf(snd_ecdf, non_zero_sample)
mod_sample = modified_arg(non_zero_sample)
```

Постройте график (`plt.scatter`) для первого способа красным цветом, а для второго — синим. Не забудьте про `alpha=0.05` и легенду. Такой график называется `Weibull plot` и является аналогом `qqplot` для распределения Вейбулла.

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

axs[0].scatter(mod_sample, mod_first_ecdf_val, c='r', alpha=0.05, 
                label='First weib. plot')
axs[0].scatter(mod_sample, mod_second_ecdf_val, c='b', alpha=0.05, 
                label='Second weib. plot')

axs[1].scatter(mod_sample, mod_first_ecdf_val / mod_sample, 
                c='r', alpha=0.03, label='First weib. plot')
axs[1].scatter(mod_sample, mod_second_ecdf_val / mod_sample, 
                c='b', alpha=0.03, label='Second weib. plot')

first_leg = axs[0].legend()
for lh in first_leg.legendHandles: 
    lh.set_alpha(1)
    
second_leg = axs[1].legend()
for lh in second_leg.legendHandles: 
    lh.set_alpha(1)
    
axs[1].set_ylim((-2, 2))
plt.show()
```

Логарифм функции правдоподобия для распределения Вейбулла:  
$L_\gamma(X) = n\ln(\gamma) + (\gamma - 1) \sum\limits_{i=1}^n \ln(X_i) - \sum\limits_{i=1}^n X_i^\gamma$

```python
my_1d_power(np.array([2, 4, 1]), np.array([3, 2]))
```

```python
DAYS_IN_FOUR_YEARS = 365 * 4 + 1

param_grid = np.logspace(-2, 2, int(1e4))

def my_1d_power(base, index):
    """
    base : numpy array of shape (`m`,)
    index : numpy array of shape (`n`,)

    Return : numpy array of shape (`m`) if m == n or n == 1
             otherwise, return numpy array of shape ('n', 'm'), which consists of 
             all possible pairwise exponentiations
    """
    if type(index) != np.ndarray:
        index = np.array(index)
        
    if (base.size == index.size or index.size == 1):
        return np.power(base, index)
    
    repeated_base = np.tile(base[np.newaxis, :], (index.size, 1))
    return np.power(repeated_base, index[:, np.newaxis])


def weibull_ll_func(sample, form_param):
    size = sample.size
    return size * np.log(form_param) + \
            + (form_param - 1) * np.sum(np.log(sample)) - \
            np.sum(my_1d_power(sample, form_param), axis=-1)
```

```python
limited_sample = weibull_sample[:DAYS_IN_FOUR_YEARS]
limited_sample = limited_sample[limited_sample > 0]
lim_samp_best_ind = weibull_ll_func(limited_sample, param_grid).argmax()
lim_samp_best_param = param_grid[lim_samp_best_ind]

full_samp_best_ind = weibull_ll_func(non_zero_sample, param_grid).argmax()
full_samp_best_param = param_grid[full_samp_best_ind]

print('ОМП параметра формы за 4 года = %lf' % lim_samp_best_param)
print('ОМП параметра формы за 10 лет = %lf' % full_samp_best_param)
```

## Задача 3

Сгенерируйте выборки $X_1, \ldots, X_N$ из $\mathit{N}(0, \theta)$, $U(0, \theta)$, $\Gamma(1, \theta)$ (**параметризация $k, \theta$**), $\theta = 3$ для всех распределений $(N = 1000).$ Для всех $n \leq N$ посчитайте значения оценок (по выборке $X_1, \ldots X_n$) методом моментов. Постройте график ошибки оценки от реального значения ($|\hat\theta - \theta|_{l_1}$) относительно размера выборки.

```python
THETA = 3
N = 1000
NORM_MEAN = 0
NORM_VAR = THETA
UNIF_LEFT = 0
UNIF_RIGHT = THETA
GAMMA_FIRST = 1
GAMMA_SECOND = THETA

norm_sample = sps.norm(loc=NORM_MEAN, scale=np.sqrt(NORM_VAR)).rvs(size=N)
unif_sample = sps.uniform(loc=UNIF_LEFT, scale=UNIF_LEFT+UNIF_RIGHT).rvs(size=N)
gamma_sample = sps.gamma(a=GAMMA_SECOND, scale=1/GAMMA_FIRST).rvs(size=N)
```

Оценки по методу моментов:

$\xi \sim \mathcal{N}(a, \sigma^2) \Rightarrow
    \left\{
    \begin{aligned}
        & a = \overline{X} \\
        & \sigma^2 = \overline{X^2} - \overline{X}^2 \\
    \end{aligned} \right.$

$\xi \sim \mathcal{U}(a, b) \Rightarrow
    \left\{
    \begin{aligned}
        & a = \sqrt{\overline{X^2} - \frac13 \overline{X}^2} - \overline{X} \\
        & b = \sqrt{\overline{X^2} - \frac13 \overline{X}^2} + \overline{X} \\
    \end{aligned} \right.$

$\xi \sim \mathcal{Г}(\alpha, \lambda) \Rightarrow
    \left\{
    \begin{aligned}
        & \alpha = \dfrac{\overline{X}}{S^2} \\
        & \lambda = \dfrac{\overline{X}^2}{S^2} \\
    \end{aligned} \right.$

```python
def sample_n_mean(sample):
    n_arr = np.arange(1, sample.size + 1)
    return np.cumsum(sample) / n_arr
    

def sample_n_snd_mom(sample):
    n_arr = np.arange(1, sample.size + 1)
    return np.cumsum(np.square(sample)) / n_arr
    

def sample_n_var(sample):
    return sample_n_snd_mom(sample) - \
            np.square(sample_n_mean(sample))


norm_first_estim = sample_n_mean(norm_sample)
norm_snd_estim = sample_n_var(norm_sample)

unif_n_mean = sample_n_mean(unif_sample)
unif_pseudo_n_std = np.sqrt(sample_n_snd_mom(unif_sample) - \
                          np.square(unif_n_mean) / 3)
unif_first_estim = unif_pseudo_n_std - unif_n_mean
unif_snd_estim = unif_pseudo_n_std + unif_n_mean

gamma_n_mean = sample_n_mean(gamma_sample)
gamma_n_var = sample_n_var(gamma_sample)
gamma_n_var[0] = 1 # avoid division by zero error
gamma_first_estim = gamma_n_mean / gamma_n_var
gamma_snd_estim = np.square(gamma_n_mean) / gamma_n_var
```

```python
N_arr = np.arange(1, N + 1)

fig, ax = plt.subplots(3, 1, figsize=(10, 18), sharex=False)

ax[0].plot(N_arr, np.abs(norm_first_estim - NORM_MEAN), c='b', 
            label='$a$ deviation')
ax[0].plot(N_arr, np.abs(norm_snd_estim - NORM_VAR), c='g', 
            label='$\sigma^2$ deviation')
ax[0].set_xlim((50, N))
ax[0].set_ylim((0, 0.5))
ax[0].set_title('Method of moments for '
                '$\\xi \\sim \\mathcal{N}(a, \\sigma^2)$')
ax[0].legend()


ax[1].plot(N_arr, np.abs(unif_first_estim - UNIF_LEFT), c='b', 
            label='$a$ deviation')
ax[1].plot(N_arr, np.abs(unif_snd_estim - UNIF_RIGHT), c='g', 
            label='$b$ deviation')
ax[1].set_ylim((0, 0.2))
ax[1].set_xlim((50, N))
ax[1].set_title('Method of moments for $\\xi \\sim \\mathcal{U}(a, b)$')
ax[1].legend()


ax[2].plot(N_arr, np.abs(gamma_first_estim - GAMMA_FIRST), c='b', 
            label='$\\alpha$ deviation')
ax[2].plot(N_arr, np.abs(gamma_snd_estim - GAMMA_SECOND), c='g', 
            label='$\\lambda$ deviation')
ax[2].set_ylim((0, 0.5))
ax[2].set_xlim((50, N))
ax[2].set_title('Method of moments for '
                '$\\xi \\sim \\mathcal{Г}(\\alpha, \\lambda)$')
ax[2].legend()

plt.show()
```

**Вывод:**  
На практике убедились в сильной состоятельности оценок по методу моментов

### Бутстреп

Для реальных данных часто сложно подобрать распределение и нужную параметризацию относительно $\theta$. Кроме того на практике сложно посчитать дисперсию оценки (для этого хотя бы нужно знать распределение, из которого пришла выборка). На помощь в таких случаях приходит **бутстреп**.

Идея очень простая. Давайте возьмем нашу выборку размера $N$ и сгенерируем из нее еще $K$ выборок. Более формально для каждой бутстрепной выборки N раз будем выбирать элементы из исходной выборки с возвращением. Полученная таким образом выборка будет содержать $\approx 63$% уникальных элементов, но это не страшно. Для всех $K$ выборок посчитаем оценку $\hat\theta$. Таким образом мы получим $K$ оценок параметра. Можно показать, что если размер бутстрепных выборок и исходной совпадают, то оценка дисперсии $s^2(\hat\theta)$, полученная из $K$ оценок, будет *хорошей*.

        \sigma^2 = \overline{X^2} - \overline{X}^2
$

```python
K_arr = np.array([10, 50, *np.arange(100, 1001, 100)]) 
bs_means = []
bs_stds = []

for K in K_arr:
    K_ind = np.random.randint(0, K, K**2)
    bs_sample = norm_sample[K_ind].reshape(K, K)
    norm_var_estim = np.mean(bs_sample**2, axis=1) - \
                        np.mean(bs_sample, axis=1)**2
    bs_means.append(norm_var_estim.mean())
    bs_stds.append(norm_var_estim.std())

bs_means = np.array(bs_means)
bs_stds = np.array(bs_stds)

plt.figure(figsize=(11, 6))
plt.plot(K_arr, bs_means, c='r', label='Mean value')
plt.fill_between(K_arr, bs_means-bs_stds, bs_means+bs_stds, 
                    facecolor='r', alpha = 0.2, 
                    label='+- sigma range')

plt.xlabel('Bootstrap sample size')
plt.ylabel('Value')
plt.title('Bootstrap sample stats for '
            '$\\xi \\sim \\mathcal{N}(0, \\theta)$')
plt.legend(loc='lower right')
plt.show()
```

$\xi \sim \mathcal{U}(a, b) \Rightarrow
        b = \sqrt{\overline{X^2} - \frac13 \overline{X}^2} + \overline{X} \\
$

```python
def get_unif_right_estim(samples):
    means = np.mean(samples, axis=1)
    pseudo_std = np.sqrt(np.mean(np.square(samples), axis=1) - \
                            np.square(means)/3)
    return pseudo_std + means
    

bs_means = []
bs_stds = []

for K in K_arr:
    K_ind = np.random.randint(0, K, K**2)
    bs_sample = unif_sample[K_ind].reshape(K, K)
    unif_right_estim = get_unif_right_estim(bs_sample)
    bs_means.append(unif_right_estim.mean())
    bs_stds.append(unif_right_estim.std())

bs_means = np.array(bs_means)
bs_stds = np.array(bs_stds)

plt.figure(figsize=(11, 6))
plt.plot(K_arr, bs_means, c='r', label='Mean value')
plt.fill_between(K_arr, bs_means-bs_stds, bs_means+bs_stds, 
                    facecolor='r', alpha = 0.2, 
                    label='+- sigma range')

plt.xlabel('Bootstrap sample size')
plt.ylabel('Value')
plt.title('Bootstrap sample stats for '
            '$\\xi \\sim \\mathcal{U}(0, \\theta)$')
plt.legend(loc='lower right')
plt.show()
```

$\xi \sim \mathcal{Г}(\alpha, \lambda) \Rightarrow
        \lambda = \dfrac{\overline{X}^2}{S^2} \\
$

```python
def get_gamma_snd_estim(samples):
    means = samples.mean(axis=1)
    samples_var = np.var(samples, axis=1)
    return np.square(means) / samples_var


bs_means = []
bs_stds = []

for K in K_arr:
    K_ind = np.random.randint(0, K, K**2)
    bs_sample = gamma_sample[K_ind].reshape(K, K)
    gamma_snd_estim = get_gamma_snd_estim(bs_sample)
    bs_means.append(gamma_snd_estim.mean())
    bs_stds.append(gamma_snd_estim.std())
    
bs_means = np.array(bs_means)
bs_stds = np.array(bs_stds)

plt.figure(figsize=(11, 6))
plt.plot(K_arr, bs_means, c='r', label='Mean value')
plt.fill_between(K_arr, bs_means-bs_stds, bs_means+bs_stds, 
                    facecolor='r', alpha = 0.2, 
                    label='+- sigma range')

plt.xlabel('Bootstrap sample size')
plt.ylabel('Value')
plt.title('Bootstrap sample stats for '
            '$\\xi \\sim \\mathcal{Г}(1, \\theta)$')
plt.legend()
plt.show()
```
